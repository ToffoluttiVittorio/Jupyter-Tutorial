## First, the packages

*   io from skimage
*   numpy
*   matplotlib



```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
```

## Then we downloads the images and define the output image

```python
# Downlaod the irc and segmentation image 

from skimage import io

irc_image_url = "https://raw.githubusercontent.com/Vitto50/Jupyter-Tutorial/main/ImageIRC.tif"
segmentation_image_url = "https://raw.githubusercontent.com/Vitto50/Jupyter-Tutorial/main/IRC_Segmentation.tif"

irc_image = io.imread(irc_image_url)
segmentation_image = io.imread(segmentation_image_url)
mean_irc_image = np.zeros_like(irc_image)

plt.imshow(irc_image)
plt.title("IRC image")
plt.axis('off')
plt.show()

```

## Do the treatment

```python
# Get unique segmentation_image values
unique_segments = np.unique(segmentation_image)

# For each segment
for segment in unique_segments:
    # Extract pixels belonging to the segment
    segment_pixels = irc_image[np.where(segmentation_image == segment)]
    
    # Calculate the average for each channel
    mean_r = np.mean(segment_pixels[:, 0])
    mean_g = np.mean(segment_pixels[:, 1])
    mean_b = np.mean(segment_pixels[:, 2])
    
    # Reassign values for each channel in the output image
    mean_irc_image[np.where(segmentation_image == segment)] = [mean_r, mean_g, mean_b]
```

## Display the output data

```python
# Save output image
# io.imsave('mean_irc_image.tif', mean_irc_image)

# Display segmented image
plt.imshow(mean_irc_image)
plt.title("Segmented image")
plt.axis('off')
plt.show()
```
